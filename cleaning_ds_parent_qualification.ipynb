````python
qualification_dict = {"Grouping": [8, 8, 7, 7, 6, 5, 5, 4, 4, 4, 4, 4, 3, 3, 3, 3, 3, 3, 2, 2, 2, 2, 2, 2, 2, 1, None, None],
                     "Qualifcation": [5, 44, 4, 43, 2, 3, 40, 22, 18, 39, 41, 42, 1, 9, 12, 10, 14, 19, 26, 11, 27, 29, 30, 38, 37, 35, 6, 34]}
qual_df = pd.DataFrame(qualification_dict)
qual_df 
````

Merging the original dataframe with the grouped dataframe

````python
mother_qual_df = pd.merge(my_df, qual_df, how="inner", left_on="Mother's qualification", right_on="Qualifcation")
mother_qual_df = mother_qual_df.drop(["Mother's qualification", "Qualifcation"], axis=1)  # dropping the detailed qualifcations
mother_qual_df.rename(columns={"Grouping": "Mother's Qualification"}, inplace=True) 


ds_final_df = pd.merge(mother_qual_df, qual_df, how="inner", left_on="Father's qualification", right_on="Qualifcation")
ds_final_df = ds_final_df.drop(["Father's qualification", "Qualifcation"], axis=1)  # dropping the detailed qualifcations
ds_final_df.rename(columns={"Grouping": "Father's Qualification"}, inplace=True)

ds_final_df["Parent's Highest Qualifcaiton"] = np.maximum(ds_final_df["Mother's Qualification"],
                                                          ds_final_df["Father's Qualification"])  # ensuring we keep NaNs
ds_final_df
````
